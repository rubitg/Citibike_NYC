```python
import pandas as pd
```

```python
cb_201309=pd.read_csv("201309 - Citi Bike trip data.csv")
cb_201306=pd.read_csv("201306 - Citi Bike trip data.csv")
cb_201312=pd.read_csv("2013-12 - Citi Bike trip data.csv")
```

```python
#converto to Dataframe
cb_201309_DF=pd.DataFrame(cb_201309)
cb_201306_DF=pd.DataFrame(cb_201306)
cb_201312_DF=pd.DataFrame(cb_201312)
```

```python
#random sample 5%
RS5_201309 = cb_201309_DF.sample(frac=0.05)
RS5_201306 = cb_201306_DF.sample(frac=0.05)
RS5_201312 = cb_201312_DF.sample(frac=0.05)

RS_2013=pd.concat([RS5_201309,RS5_201306,RS5_201312])
RS_2013.head()
RS_2013.describe()
```

```python
#copy to csv
RS_2013.to_csv('RS_2013.csv')
```
